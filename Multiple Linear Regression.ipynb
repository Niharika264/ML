```python
# Data Wrangling --> Cleaning the data
df.isna().sum()
```

```python
# car_id and car_name are non-imp features hence they can be dropped
df.drop('car_ID',inplace=True,axis=1)
```

```python
df.drop('CarName',inplace=True,axis=1)
```

```python
df.head(10)
```

```python
df
```

```python
df['fueltype_gas']=pd.get_dummies(df['fueltype'],drop_first=True)
df['aspiration_std']=pd.get_dummies(df['aspiration'],drop_first=True)
df['doornumber_two']=pd.get_dummies(df['doornumber'],drop_first=True)
df['enginelocation_front']=pd.get_dummies(df['enginelocation'],drop_first=True)
```

```python
#Unique values in carbody=['convertible' 'hatchback' 'sedan' 'wagon' 'hardtop']
my_dict = {'convertible':0,'hatchback':1,'sedan':2,'wagon':3,'hardtop':4}
df['carbody_encoded']=df['carbody'].map(my_dict)
```

```python
df['carbody_encoded']
```

```python
#Unique values in drivewheel=['rwd' 'fwd' '4wd']
my_dict = {'rwd':0,'fwd':1,'4wd':2}
df['drivewheel_encoded']=df['drivewheel'].map(my_dict)
```

```python
#Unique values in enginetype=['dohc' 'ohcv' 'ohc' 'l' 'rotor' 'ohcf' 'dohcv']
my_dict = {'dohc':0,'ohcv':1,'ohc':2,'l':3,'rotor':4,'ohcf':5,'dohcv':6}
df['enginetype_encoded']=df['enginetype'].map(my_dict)
```

```python
#Unique values in cylindernumber=['four' 'six' 'five' 'three' 'twelve' 'two' 'eight']
my_dict = {'four':4,'six':6,'five':5,'three':3,'twelve':12,'two':2,'eight':8}
df['cylinder_no_encoded']=df['cylindernumber'].map(my_dict)
```

```python
#Unique values in fuelsystem=['mpfi' '2bbl' 'mfi' '1bbl' 'spfi' '4bbl' 'idi' 'spdi']
my_dict = {'mpfi':0,'2bbl':1,'mfi':2,'1bbl':3,'spfi':4,'4bbl':5,'idi':6,'spdi':7}
df['fuelsys_encoded']=df['fuelsystem'].map(my_dict)
```

```python
df
```

```python
df.drop('fueltype',axis=1,inplace=True)
df.drop('aspiration',axis=1,inplace=True)
df.drop('doornumber',axis=1,inplace=True)
df.drop('carbody',axis=1,inplace=True)
df.drop('drivewheel',axis=1,inplace=True)
df.drop('enginelocation',axis=1,inplace=True)
df.drop('enginetype',axis=1,inplace=True)
df.drop('cylindernumber',axis=1,inplace=True)
df.drop('fuelsystem',axis=1,inplace=True)
```

```python
## Feature selection:
#based on correlation Coefficients--> for numerical feature columns
X = df.iloc[:,[1,2,3,5,6,7,10,12,13,15,16,17,18,19,20,21,22,23]].values
```

```python
X
```

```python
Y = df.iloc[:,[14]].values
```

```python
Y
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X = sc.fit_transform(X)
X
```

```python
# splitting into training and testing data
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X_tr,X_ts,Y_tr,Y_ts = train_test_split(X,Y,random_state=0,test_size=0.3)
```

```python
X_tr.shape
```

```python
Y_tr.shape
```

```python
# Model building
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# fitting the training data
model.fit(X_tr,Y_tr)
```

```python
y_pred=model.predict(X_ts)
```

```python
r2_score(Y_ts,y_pred)
```

```python
plt.scatter(x=range(len(Y_ts)),y=Y_ts,label='actual',c='g')
plt.scatter(x=range(len(y_pred)),y=y_pred,label='predicted',c='b')
plt.legend()
plt.show()
```
