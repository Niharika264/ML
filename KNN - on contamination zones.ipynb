```python
X = df3.iloc[:,0:-1].values
Y = df3.iloc[: ,[-1]].values
```

```python
X
```

```python
#Y
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
Y.shape
```

```python
# Y is a vector but 1-D array is expected
Y=Y.reshape(-1)
```

```python
Y.shape
```

```python
#Y
```

```python
# splitting into train and test data
from sklearn.model_selection import train_test_split
```

```python
X_tr,X_ts,Y_tr,Y_ts = train_test_split(X,Y,random_state=0,test_size=0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
# selecting k value
accuracy = []
error = []
for i in range(1,25):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_tr,Y_tr)
    Y_pred = model.predict(X_ts)
    acc = accuracy_score(Y_ts,Y_pred)
    #print("For k = {}, accuracy = {}".format(i,acc))
    accuracy.append(acc)
    error.append(1-acc)
```

```python
plt.plot(list(range(1,25)),accuracy)
plt.xlabel('K value')
plt.ylabel('accuracy rate')
plt.show()
```

# From the graph the accuracy is max when k = 23 

```python
# Model Building

model = KNeighborsClassifier(n_neighbors=23)
model.fit(X_tr,Y_tr)
Y_pred = model.predict(X_ts)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=2)
```

```python
print("Before SMOTE:\nlabel 0: {}".format(sum(Y_tr==0)))
print("label 1: {}".format(sum(Y_tr==1)))
print("label 2: {}".format(sum(Y_tr==2)))
```

```python
print("After SMOTE:\n label 0: {}".format(sum(Y_tr_smote==0)))
print("label 1 :{}".format(sum(Y_tr_smote==1)))
print("label 2 :{}".format(sum(Y_tr_smote==2)))
```
